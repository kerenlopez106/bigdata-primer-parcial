# Ejercicio 16 DataFrame de estudiantes

```python
import pandas as pd

datos_estudiantes = {
    'nombre': ['Carlos', 'Ana', 'Luis', 'Sofia', 'Fernando', 'Maria', 'Jorge', 'Elena', 'Ricardo', 'Beatriz'],
    'carrera': ['Sistemas', 'Medicina', 'Sistemas', 'Derecho', 'Industrial', 'Medicina', 'Sistemas', 'Derecho', 'Industrial', 'Sistemas'],
    'edad': [21, 23, 20, 22, 24, 21, 22, 20, 25, 23],
    'promedio': [85.5, 92.0, 78.0, 88.5, 65.0, 90.0, 81.5, 74.0, 89.0, 95.5],
    'ciudad': ['Tegucigalpa', 'San Pedro Sula', 'Tegucigalpa', 'La Ceiba', 'Choluteca', 'San Pedro Sula', 'Comayagua', 'Tegucigalpa', 'La Ceiba', 'Santa Rosa']
}

df = pd.DataFrame(datos_estudiantes)

print("--- 3. Muestre las primeras 5 filas ---")
display(df.head(5)) 

print("\n--- 4. Muestre la información general del DataFrame ---")
df.info()

print("\n--- 5. Calcule el promedio general de calificaciones ---")
promedio_general = df['promedio'].mean()
print(f"El promedio general de todos los estudiantes es: {promedio_general:.2f}")

print("\n--- 6. Muestre estudiantes con promedio mayor o igual a 80 ---")
estudiantes_altos = df[df['promedio'] >= 80]
display(estudiantes_altos)

print("\n--- 7. Ordene estudiantes por promedio de mayor a menor ---")
df_ordenado = df.sort_values(by='promedio', ascending=False)
display(df_ordenado)
```

# Ejercicio 17 Análisis de ventas con pandas


```python
import pandas as pd

# Creamos las 15 ventas
datos_ventas = {
    'Fecha': ['2026-05-01', '2026-05-02', '2026-05-03', '2026-05-04', '2026-05-05', 
              '2026-05-06', '2026-05-07', '2026-05-08', '2026-05-09', '2026-05-10', 
              '2026-05-11', '2026-05-12', '2026-05-13', '2026-05-14', '2026-05-15'],
    'Vendedor': ['Carlos', 'Ana', 'Luis', 'Ana', 'Carlos', 
                 'Luis', 'Ana', 'Carlos', 'Luis', 'Ana', 
                 'Carlos', 'Luis', 'Ana', 'Carlos', 'Luis'],
    'Producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Mouse', 
                 'Laptop', 'Teclado', 'Monitor', 'Laptop', 'Mouse', 
                 'Teclado', 'Monitor', 'Laptop', 'Teclado', 'Monitor'],
    'Categoria': ['Tecnología', 'Accesorios', 'Accesorios', 'Tecnología', 'Accesorios', 
                  'Tecnología', 'Accesorios', 'Tecnología', 'Tecnología', 'Accesorios', 
                  'Accesorios', 'Tecnología', 'Tecnología', 'Accesorios', 'Tecnología'],
    'Cantidad': [1, 5, 2, 1, 3, 2, 4, 1, 1, 6, 3, 2, 1, 2, 1],
    'Precio Unitario': [800, 20, 35, 150, 20, 800, 35, 150, 800, 20, 35, 150, 800, 35, 150]
}

# Convertimos los datos en una tabla
df = pd.DataFrame(datos_ventas)


df
```

```python

df['TotalVenta'] = df['Cantidad'] * df['Precio Unitario']


df
```

```python

print("Total vendido por vendedor")
total_por_vendedor = df.groupby('Vendedor')['TotalVenta'].sum()
print(total_por_vendedor)
print("\n")

print("Producto más vendido")
producto_mas_vendido = df.groupby('Producto')['Cantidad'].sum().idxmax()
print(f"El producto del que se vendieron más unidades es: {producto_mas_vendido}")
print("\n")

print("Categoría con mayor ingreso")
categoria_mayor_ingreso = df.groupby('Categoria')['TotalVenta'].sum().idxmax()
print(f"La categoría que generó más dinero es: {categoria_mayor_ingreso}")
print("\n")

print("Ventas ordenadas de mayor a menor")
df_ordenado = df.sort_values(by='TotalVenta', ascending=False)
df_ordenado
```

# Ejercicio 18 Limpieza de datos con pandas

```python
# Ejercicio 18: Limpieza de datos con pandas
import pandas as pd
clientes = [
    {"nombre": "Tommy Vega", "edad": 30, "email": None, "ciudad": "San Pedro Sula", "telefono": "98765432"},
    {"nombre": "Nelsy Alonzo", "edad": 24, "email": "nelsy.alonzo@gmail.com", "ciudad": "Puerto Cortés", "telefono": None},
    {"nombre": "Marlon Rivera", "edad": 29, "email": "marlon.rivera@gmail.com", "ciudad": None, "telefono": "92366718"}
]

df_clientes = pd.DataFrame(clientes)

print(df_clientes.isnull().sum())

print("\nValores nulos:", df_clientes[df_clientes.isnull().any(axis=1)].shape[0])

df_clientes["ciudad"] = df_clientes["ciudad"].fillna("No especificada")

df_limpio = df_clientes.dropna(subset=["email"])

display(df_limpio)

# La lógica en este ejercicio es crear primero un diccionario de los clientes y luego pasarlo a un DataFrame, posterior a ello se solicitan varias cosas
# Con la función isnull() se identifican lso valores nulos, y luego solo se cuentan con la funcion de sum()
# Luego para contar los valores nulos por columna se utiliza la misma funcion de isnull() pero con un filtro para contar solo las filas que tengan al menos un valor nulo
# Ya con la funciona fillna() se reemplaza el valor nulo de la columna ciudad por "No especificada"
# Y por último con la función dropna() se eliminan las filas que tengan un valor nulo en la columna email
```

# Ejercicio 19

```python
import pandas as pd
datos_prestamos = {
    "cliente":         ["Ana Torres", "Luis Mejia", "Maria Paz", "Carlos Vega",
                        "Sofia Ruiz", "Jorge Lima", "Valeria Cruz", "Diego Reyes",
                        "Fernanda Lopez", "Andres Diaz"],
    "monto_solicitado":[45000, 75000, 30000, 120000, 55000,
                        20000, 90000, 48000, 65000, 15000],
    "plazo_meses":     [24, 36, 12, 48, 24, 12, 36, 24, 36, 12],
    "ingreso_mensual": [18000, 25000, 12000, 40000, 20000,
                        9000, 35000, 17000, 22000, 8000],
    "estado":          ["Aprobado", "Aprobado", "Rechazado", "Aprobado",
                        "En revision", "Rechazado", "Aprobado", "En revision",
                        "Aprobado", "Rechazado"]
}

df_pre = pd.DataFrame(datos_prestamos)
print("DataFrame de prestamos:")
print(df_pre)

print(f"\nMonto promedio solicitado: L.{df_pre['monto_solicitado'].mean():,.2f}")

print("\nConteo de solicitudes por estado:")
print(df_pre["estado"].value_counts())

print("\nPrestamos aprobados:")
print(df_pre[df_pre["estado"] == "Aprobado"])

print("\nClientes con monto mayor a L.50,000:")
print(df_pre[df_pre["monto_solicitado"] > 50000])

print("\nPromedio de monto solicitado por estado:")
print(df_pre.groupby("estado")["monto_solicitado"].mean().reset_index())
```

# Ejercicio 20 Mini analisis de datos

```python
# Ejercicio 20: Mini análisis de datos tipo empresa
import pandas as pd
empleados = [
    {"nombre": "Tommy Vega", "departamento": "IT", "cargo": "Analista de Sistemas", "salario": 33000, "años_experiencia": 7, "ciudad": "San Pedro Sula"},
    {"nombre": "Nelsy Alonzo", "departamento": "Operaciones", "cargo": "Especialista en Digitación", "salario": 18000, "años_experiencia": 2, "ciudad": "Puerto Cortés"},
    {"nombre": "Marlon Rivera", "departamento": "IT", "cargo": "Desarrollador Senior", "salario": 42000, "años_experiencia": 6, "ciudad": "San Pedro Sula"}
 ]

df_empleados = pd.DataFrame(empleados)
display(df_empleados)

print("\nEstadísticas generales de salarios")
estadisticas_salarios = df_empleados["salario"].describe()
display(estadisticas_salarios)

print("\nSalario promedio por departamento")
promedio_por_depto = df_empleados.groupby("departamento")["salario"].mean()
display(promedio_por_depto)

print("\nEmpleado con salario más alto")
top_empleado = df_empleados.loc[df_empleados["salario"].idxmax()]
display(top_empleado)

print("\nEmpleados con más de 5 años de experiencia")
df_mas_experiencia = df_empleados[df_empleados["años_experiencia"] > 5]
display(df_mas_experiencia)

print("\nEmpleados ordenados por salario")
df_ordenado = df_empleados.sort_values(by="salario", ascending=False)
display(df_ordenado)

def nivel_experiencia(años):
    if años < 1:
        return "Junior"
    elif años <= 3:
        return "Intermedio"
    else:
        return "Senior"

df_empleados["NivelExperiencia"] = df_empleados["años_experiencia"].apply(nivel_experiencia)
display(df_empleados)

# La lógica en este ejercicio de igual forma se basa en crear un diccionario primero y luego convertirlo a un DataFrame
# Luego con la funcion de describe() se pueden obtener las estadisticas generales de la columna de salarios
# Para el salario promerio por departamente se utiliza primero la funcion groupby() y luego se obtiene el promedio con la funcion mean()
# Para obtener el empleado con el salario mas alto se utilia la funcion idxmx() que nos permite por medio del id obtener el salario mas alto de X empleado
# y utilizando la funcion loc() se obttiene toda la fila del empleado
# La del empeado con mas de 5 años de experiencia simplemente se evalua la columan de años de experiencia que sea mayor a 5
# Para ordenar los empleados por salario, se utiliza la funcion sort_values() y se ordena de forma descendente
# Y ya por último creamos primero la defincion para saber que se considera Junior, Interemedio o Senior, luego ya con la funcion apply() se aplica esa función 
# a la columna de años de experiencia para crear una nueva columna con el nivel de experiencia de cada empleado
```
