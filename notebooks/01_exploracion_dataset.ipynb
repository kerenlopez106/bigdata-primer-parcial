## Caso de aplicacion: Netflix- La personalización y creación de contenidos de Netflix

### ¿Que datos genera?
<p>Con más de 301,63 millones suscriptores en todo el mundo, Netflix es uno de los ejemplos de cómo el 'big data' puede favorecer el éxito de una plataforma de 'streaming': recopila y analiza las interacciones de los usuarios, rastreando detalles como los hábitos de visionado, las pausas y las valoraciones de los usuarios para tomar decisiones que van desde la recomendación de contenidos hasta la creación de programación </p>

<p>Netflix dispone de un apartado en su página web en el que registra las películas y series más vistas por país y fecha. Globalmente, las diez series de habla inglesa más vistas de su plataforma son producciones originales de la compañía. Una muestra más de la importancia que tiene el análisis de las preferencias de sus usuarios para ofrecer nuevas series.</p>


### ¿Quien los usa?
<p>Los algoritmos de recomendacion de Netflix,los ingenieros de datos y directores de contenido  de esta manera el equipo decide que series y contenido producir segun la informacion que recopilan.</p> 

### ¿Que problema resuelve?
<p>Permite hacer inversiones con garantia que la perdida monetaria será minima ya que toman como base los datos recopilados. Asi mismo garantiza una retención de usuarios ya que mantienen contenido que se asemeja mucho a sus gustos y preferencias o programas vistos anteriormente. Permite brindar un servicio mas personalizado a cada usuario, lo cual beneficia a la empresa en varios aspectos.</p>

### ¿Porque no bastaría Excel? 
<p>Excel no bastaría ya que no es capaz de manejar ese volumen masivo de datos, ademas seria ineficiente para poder hacer un analisis acertado sobre el comportamiento de millones de usuarios. No es capaz de ejecutar los algoritmos de prediccion que maneja Netflix y no cuenta con la velocidad adecuada para el procesamiento de la informacion. Seria ineficiente.</p>

### Justificación 
<p>Se considera que es Big Data debido al volumen y la velocidad del procesamiento de la información de millones de usuarios y su comportamiento dentro de la plataforma. El manejo masivo de datos de Netflix podria colapsar cualquier otro tipo de almacenamiento de datos como una base de datos tradicional, además cumple con las 4Vs de Big data: velocidad, volumen, veracidad y variedad.  </p>
