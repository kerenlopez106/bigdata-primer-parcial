# Ejercicio 1: Clasificación de edad
**Enunciado:** determinar si una persona es niño, adolescente, aduto o adulto mayor segun su edad.
**Lógica:** se evalua la edad con condiciones if, elif y else. El primer rango que se cumple dtermina la categoría.

```python
def clasificar_edad(edad):
    if edad < 13:
        return "Niño"
    elif edad <= 17:
        return "Adolescente"
    elif edad <= 59:
        return "Adulto"
    else:
        return "Adulto mayor"

edad = 60
categoria = clasificar_edad(edad) 
print(f"Edad: {edad} años")
print(f"Categoría: {categoria}")



```

# Ejercicio 2: Conversor de temperatura 
**Enunciado:** Conviertir una temperatura de grados Celsius a Fahrenheit.
**Lógica:** Se crea una función y recibe el valor en Celsius, luego se aplica una operación aritmetica y regresa el valor calculado en Fahrenheit.


```python
def convertirt(celsius):
    fahrenheit = (celsius * 9 / 5) + 32
    return fahrenheit

celsius = 15 
resultadof = convertirt(celsius)
print(f"Temperatura: {celsius}")
print(f"Resultado:  {resultadof} °F")
    
```

# Ejercicio 3 Calculadora básica
**Enunciado:** Defina dos números y una operación matemática
**Lógica:** Se definieron dos variables numéricas. Definimos un proceso para cada simbolo matematico.

```python
num1 = 33
num2 = 1
operacion = "suma"  # Puede ser: suma, resta, multiplicacion, division

if operacion == "suma":
    resultado = num1 + num2
    print(f"Resultado: {num1} + {num2} = {resultado}")
elif operacion == "resta":
    resultado = num1 - num2
    print(f"Resultado: {num1} - {num2} = {resultado}")
elif operacion == "multiplicacion":
    resultado = num1 * num2
    print(f"Resultado: {num1} * {num2} = {resultado}")
elif operacion == "division":
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado: {num1} / {num2} = {resultado}")
    else:
        print("Error: No se puede dividir entre cero.")
else:
    print("Operación no permitida.")
```

# Ejercicio 6: Contador de vocales
**Enunciado:** Defina una frase y cuente cuantas vocales tiene
**Lógica:** Se define la cadena de texto y se utiliza un ciclo for para recorrer los caracteres de la frase. Para que no importe si es mayuscula o  minuscula se hace uso de .lower

```python
frase = "La practica hace al maestroo"
vocales = "aeiou"
contador = 0

for letra in frase.lower():
    if letra in vocales:
        contador += 1

print(f"Frase: {frase}")
print(f"Cantidad de vocales: {contador}")
```

# Ejercicio 7 Promedio de calificaciones
**Enunciado:** Defina 5 calificaciones y saque el promedio, ademas mestre si el estudiante aprobo o reprobo.
**Lógica:** Se almacena la lista de las 5 notas. con sum() se suma el totaly se divide entre la cantidad de notas con len() para poder calcular el promedio. Mediante if-else se evalua si aprueba o no.

```python
calificaciones = [85, 72, 90, 68, 77]

total = sum(calificaciones)
promedio = total / len(calificaciones)

print(f"Calificaciones: {calificaciones}")
print(f"Suma total: {total}")
print(f"Promedio: {promedio:.1f}")

if promedio >= 70:
    print("Estado: Aprobado")
else:
    print("Estado: Reprobado")
```

# Ejercicio 11: Función para calcular descuento
**Enunciado:** Crear una función que reciba el precio original y el porcentaje de descuento para devolver el precio final.
**Lógica:** La función tiene el costo del producto y el porcentaje de descuento como parametros. Luego se calcula el monto dividiendo el porcentaje de descuento entre 100 y multiplicandolo por el precio. Al precio del producto se le resta el monto del descuento para saber el precio a pagar.

```python
def calcular(precio, porcentaje):
    monto = precio * ( porcentaje /100)
    precio_final = precio - monto 
    return precio_final 
precio_inicial = 700
descuento= 10
pago= calcular (precio_inicial, descuento)
print(f"Precio original: Lps. {precio_inicial}")
print(f"Descuento: {descuento}%")
print(f"Precio final: Lps.{pago}")
```

# Ejercicio 12 Validación de contraseña
**Enunciado:** Valide una contraseña definida por el usuario o por el código. Indique si la contraseña es válida o no.

**Lógica:** Se define una cadena de texto con los caracteres a evaluar y por medio de la funcion len() verifica si cumple con el tamaño minimo de caracteres. Se utiliza tambien any() .isdigit(), .isupper, .islowe() para corroborar que cumple con los requerimientos.

```python
contrasena = "Segura123"

tiene_longitud = len(contrasena) >= 8
tiene_numero = any(c.isdigit() for c in contrasena)
tiene_mayuscula = any(c.isupper() for c in contrasena)
tiene_minuscula = any(c.islower() for c in contrasena)

print(f"Contrasena: {contrasena}")
print(f"  Longitud minima (8): {tiene_longitud}")
print(f"  Contiene numero: {tiene_numero}")
print(f"  Contiene mayuscula: {tiene_mayuscula}")
print(f"  Contiene minuscula: {tiene_minuscula}")

if tiene_longitud and tiene_numero and tiene_mayuscula and tiene_minuscula:
    print("Resultado: Contrasena VALIDA")
else:
    print("Resultado: Contrasena NO valida")
```

# Ejercicio 14: Registro de ventas


```python


ventas = [
    {"cliente": "Tommy Vega", "producto": "Cinta de aislamiento", "cantidad": 10, "precio_unitario": 5.00},
    {"cliente": "Nelsy Alonzo", "producto": "Cable de red", "cantidad": 5, "precio_unitario": 100.00},
]

total_venta = 0
for venta in ventas:
    total_venta += venta["cantidad"] * venta["precio_unitario"]
    print(f"Cliente: {venta['cliente']} - Total: L {total_venta:.2f}")

total_general = 0
for venta in ventas:
    total_general += venta["cantidad"] * venta["precio_unitario"]

compra_mas_alta = 0
cliente_mas_alto = ""
for venta in ventas:
    total_venta = venta["cantidad"] * venta["precio_unitario"]
    if total_venta > compra_mas_alta:
        compra_mas_alta = total_venta
        cliente_mas_alto = venta["cliente"]
    
print(f"\nTotal general: L {total_general:.2f}")
print(f"Cliente con compra más alta: {cliente_mas_alto} - L {compra_mas_alta:.2f}\n")

# La lógica en este ejercicio se basa en crear una lista con dos ventas y luego recorrer la lista tres veces para sacar tres datos diferentes
# El primero recorre la lista para sacar el total de venta por cliente usando un ciclo for
# El segundo hace lo mismo, solamente que saca el dato de la venta general
# Ya en el tercero es un poco más completo, porque necesito sacar dos datos, la compra mas alta y el cliente de dicha compra
# Para ello, recorro la lista de ventas, voy sumando las ventas de cada uno, y luego evaluo si el total anterior es mayor o menor que la venta actual
# y almaceno el valor de la compra mas alta con su respectivo cliente en dos variables diferentes
```

# Ejercicio 15 Limpieza básica de texto
**Enunciado:** Defina o solicite una frase.
• Elimine espacios al inicio y al final.
• Convierta el texto a minúsculas.
• Reemplace espacios dobles por un solo espacio.
• Cuente cuántas palabras contiene.
• Explique los métodos strip(), lower(), replace() y split().
**Lógica:** Se define una cadena de texto con la contraseña a evaluar. Se utiliza la función len() para verificar si cumple con el tamaño mínimo de caracteres. Usa la función iterativa any()  con métodos de  Python: .isdigit() para buscar números, .isupper() para letras mayúsculas y .islower() para letras minúsculas, guardando los resultados en variables booleanas (True o False).

```python
frase_original = "   Hola  esto   es una frase de   prueba con   muchos espacios   "
print(f"Frase original: '{frase_original}'\n")

frase_limpia = frase_original.strip()

frase_limpia = frase_limpia.lower()

while "  " in frase_limpia:
    frase_limpia = frase_limpia.replace("  ", " ")

print(f"Frase limpia:   '{frase_limpia}'")

palabras = frase_limpia.split(" ")
total_palabras = len(palabras)
print(f"Número total de palabras: {total_palabras}\n")

print("="*60)
print("EXPLICACIÓN DE LOS MÉTODOS UTILIZADOS:")
print("="*60)
print("- strip(): Elimina todos los caracteres de espacio en blanco (espacios, pestañas, saltos de línea)")
print("           que se encuentren al principio (izquierda) y al final (derecha) de la cadena de texto.")
print("- lower(): Convierte todos los caracteres alfabéticos con mayúsculas de la cadena a minúsculas.")
print("- replace(antiguo, nuevo): Busca una subcadena específica dentro del texto y la reemplaza")
print("                           por otra subcadena nueva todas las veces que aparezca.")
print("- split(separador): Divide una cadena de texto en una lista de subcadenas basándose en un")
print("                    carácter separador especificado (por defecto es cualquier espacio en blanco).")
```
